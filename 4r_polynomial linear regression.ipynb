```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Position_Salaries.csv')
df.columns
```

```python
x=df[['Level']].values # matrix
y=df['Salary'].values # vector
```

```python
plt.scatter(df['Level'].values,df['Salary'].values)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression().fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=4)
```

```python
x_poly=pf.fit_transform(x)
```

```python
lr_pf=LinearRegression().fit(x_poly,y)
```

```python
# lr result visualization
```

```python
plt.scatter(x,y,color='red')
plt.plot (x,lr.predict(x),color='blue')
plt.plot(x,lr_pf.predict(x_poly),color='green')
plt.title('truth or bluff')
plt.xlabel('level')
plt.ylabel('salary')
plt.show
```

```python
# to make the curve more smothy for advanced plot
```

```python
x_grid=np.arange(min(x),max(x),0.1)
```

```python
x_grid=x_grid.reshape((len(x_grid),1))
```

```python
plt.plot(x_grid,lr_pf.predict(pf.fit_transform(x_grid)),color='green')
```
